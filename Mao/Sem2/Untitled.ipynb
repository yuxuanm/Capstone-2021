```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
reconstructed_model = keras.models.load_model("autoencoder(8,8,4) sample 3000")
```

```python
reconstructed_model
```

```python
import glob
input_data_split = []
for np_name in glob.glob('raw data/128/dataset_128_N102/*.np[yz]'):
    input_data_split.append(np.load(np_name))
input_data_split = np.array(input_data_split)
```

```python
input_data_split[1000].shape
```

```python
decoded_data = reconstructed_model.predict(np.expand_dims(input_data_split[1000], 0))
decoded_data.shape
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(16,32)) # image
ax1 = fig.add_subplot(1,2,1) 
ax1.imshow(input_data_split[1000],cmap='hot')
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(decoded_data.reshape(128,128),cmap='hot')
```
